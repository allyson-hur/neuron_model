## Neuron Basics
Neurons have four main parts:
1. **Dendrite:** receives information
2. **Cell body/soma:** processes and integrates the received information
3. **Axon:** carries information long distances from one part of the neuron to another
4. **Axon terminal:** transmits information to the next cell in the chain

A bundle of axons is called a nerve. 

## Action Potential Overview

When a neuron's action potential is reached, we say that the neuron "fires". The transmission of the neuronal signal is dependent on the movement of ions, which are charged particles. Sodium (Na), Potassium (K), and Chloride (Cl) are unevenly distributed across within the inside and outside of the cell. Na and K are both positively charged. 

When the neuron is not sending a signal, then it is at rest. At rest, there is an an ionic imabalance that refers to the **electrochemical gradient** across the membrane that consists of two parts:
1. Chemical(concentration) gradient: there is an ionic separation at the cell membrane where the concentration of Na is higher outside the cell relative to the inside, and the concentration of K is higher inside the cell relative to the outside. 
2. Electrical gradient: there is a difference in charge at the cell membrane where there are more positively charged particles outside the cell relative to the inside of the cell.
### Resting Membrane Potential
At rest, the neuronal membrane has a resting potential of around -70mV. This means that the inside of the cell is 70mV less positive than the outside of the cell. The electrical gradient contributes to establishing this potential. However, while the outside of the cell has a net positive charge and the inside of the cell has a net negative charge, the charges line up at the membrane so that the bulk solution on either side is electrically neutral. The **resting membrane potential** is the point where the cell achieves electrochemical equillibrium, meaning that the concentration gradient and the electrical gradient for each ion is equal and opposite.
### Ion Channels
Ions need a protein embedded in the membrane to facilitate movement. Most ions cross the membrane through a structure called an ion channel, of which there are three main types:
1. Voltage-gated channels: only open when the membrane potential reaches a certain value. This channel usually only lets a single ion through when open, so separate channels are needed for each ion.
2. Ligand-gated channel: open when bound by a specific molecule.
3. Mechanically-gated channels: open in response to physical forces (ie. pressure).
### Graded Potential
Most ion channels are selectively permeable. As ions move through the channels and cross the membrane, the membrane potential starts to move away from the resting potential. If the resulting change in potential is small, we call it **graded potential**, where the change is not big enough to open the voltage-gated channel. When graded potential occurs, the neuron moves to reset the potential back to resting by using the sodium-potassium pump, which uses energy generated by ATP hydrolysis to actively transport ions across the cell membrane against the concentration gradient. Sodium is transported outside the cell where its concentration is higher while potassium is transported inside the cell where its concentration is higher. This restores the electrochemical gradient back to its resting level.

### Action Potential
When stimulation at the dendrite is high enough to bring the membrane potential from its resting potential (-70mV) to its threshold potential (-55mV) or higher, this triggers the action potential. Voltage-gated channels have three states: open, closed, or inactivated. When the neuron is at rest, the Na channel is closed. Several processes happen in the following order:
1. **Depolarization:** once the cell membrane reaches its threshold potential, the channel change to an open position, which causes Na to rush into the cell due to the electrochemical gradient. As positive Na ions enter the cell, the membrane becomes less negative as it approaches 0mV.
2. **Overshoot:** the voltage gradient goes beyond 0mV up to 30mV. As the membrane potential becomes positive, the Na channel inactivation gate shuts, making the channel inactivated. This stops the flow of Na ions into the cell.
3. **Repolarization:**: the change in membrane potential opens the voltage-gated K channels. K ions flow out of the cell, making the voltage gradient less positive, then eventually negative. During this stage, the inactivated Na channels won't respond to any stimulus. The neuron is in its absolute refractory period, where a nerve cannot fire another action potential no matter how strongly itâ€™s simulated. This helps prevents action potential from traveling backwards.
4. **Hyperpolarization:** K channels are slow to close, so the membrane potential becomes more negative than its resting potential. The sodium channels are closed, and inactivation gate opens. There is no change in sodium flow, but they could be opened again. This is called relative refractory period because it would take larger than usual stimulus to reach the threshold potentialbecause the cell is hyperpolarized due to potassium still leaving the cell.

Finally, the sodium-potassium pump restablishes the potential gradient by moving more sodium ions out than potassium ions in, which returns the membrane potential back to its resting potential.

### Final Notes on Action Potential
The amplitude of action potential (max voltage in one neuron during action potential) never changes. It has an all or nothing mentailty (it either happens or doesn't happen). What can change is the frequency of action potentials. For example, the frequency of action potentials would be high when experiencing pain, but lower for feeling a breeze. Some axons do transmit action potentials faster than others, which can be due to the myelin sheaths around the axons. Myelin increases the conduction velocity, which speeds up the transmission.

Here are the main points of the action potential of a neuron.

Resting potential: no stimulus 

Graded potential: small stimulus

Action potential: stimulus is above threshold --> neuron fires!

## Hodgkin-Huxley Model

The Hodgkin-Huxley model is a mathematical model that describes the excitability of neurons with a set of nonlinear differential equations. Each component of the an excitable cell is represented as an electrical element.

### The Variables

$c_m$: capacitance that represents the lipid bilayer. The lipid bilayer is the part of the cell membrane that acts as a barrier to the passage of ions into and outside the cell.

$g_n$: electrical conductances that represent the voltage-gated ion channels. $n$ is the specific ion channel.


$g_l$: linear conductances that represent leak channels. Leak channels channel that are open even in a unsimulated or resting state (especially for potassium ions).

$E_n$: voltage sources that represent the electrochemical gradient.

$I_p$: current sources that represent the ion pumps.

$V_m$: voltage representing the membrane potential.

$V_i$: voltage representing the reversal potential of the specific ion channel. The reversal potential is the membrane voltage at which the net current through the channel is zero.


### Equations for Current Through Membrane
The current flowing through the lipid bilayer can be represented as:

$I_c = c_m\frac{dV_m}{dt}$

The current flowing through a given ion channel is the product of the channel's conductance and the driving potential for the specific ion.

$I_i=g_i(V_m-V_i)$

Combining these two equations will give us the total current through the membrane:

$I = c_m\frac{dV_m}{dt} + g_k(V_m-V_k) + g_{Na}(V_m-V_{Na})+g_l(V_m-V_l)$

$g_k$: potassium conductance per unit area.

$g_{Na}$: sodium conductance per unit area.

$V_k$: potassium reversal potential.

$V_{Na}$: sodium reversal potential.

$g_l$: leak conductance per unit area.

$V_l$: leak reversal potential.

Note: $V_m$, $g_{Na}$, and $g_k$ are all time dependent.

### Equations for Voltage-gated Ion Channels
Hodgkin and Huxley developed a model in which the properties of an excitable cell are described by a set of four ordinary differential equations. 

$I = c_m\frac{dV_m}{dt}+\bar{g_K}n^4(V_m-V_K)+\bar{g_{Na}}m^3h(V_m-V_{Na})+\bar{g_l}(V_m-V_l)$

$\frac{dn}{dt}=\alpha _n(V_m)(1-n)-\beta _n(V_m)n$

$\frac{dm}{dt}=\alpha _m(V_m)(1-m)-\beta _m(V_m)m$

$\frac{dh}{dt}=\alpha _h(V_m)(1-h)-\beta _h(V_m)h$

$I$: current per unit area.

$\bar g_n$: maximal value of the conductance.

$\alpha _i$ and $\beta _i$: rate constants for i-th ion channel.

$n$: dimensionless probability between 0 and 1 that represents the potassium channel subunit activation.

$m$: dimensionless probability between 0 and 1 that represents the sodium channel subunit activation.

$h$: dimensionless probability between 0 and 1 that represents the sodium channel subunit inactivation.
